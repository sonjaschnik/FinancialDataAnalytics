### Git (local repository)
> Git is a free and open source distributed version control system designed to handle everything from small to very large projects with speed and efficiency. [(see Git, 2023)](https://git-scm.com/)

Some source-code editors come with build-in Git (and even GitHub) capabilities or can be extended (e.g. [Microsoft's Visual Studio Code](https://code.visualstudio.com/), which I use during this course).

Your local repository is essentially a folder on your local file system. Changes made in that folder can be committed to the (local) git repository. 

First, "stage" your changes - this is sth. like a pre-commit:

```Bash
# The '*' adds all changes made in your folder (you should be more selective)
git add *
```

Second, commit your staged changes to the local repository:
```Bash
git commit -m 'Commit message'
```

In the broadest sense, you could see Git as a block chain of commits (changes) made to your repository. You can thus
* observe a complete history.
* `git checkout` the state at any commit to the repository.


More on Git:

* About Git itself: https://git-scm.com/about

### Naming convention
Here are some best practices to follow when naming stuff.
* Use all lowercase. Ex: name instead of Name
* One exception: class names should start with a capital letter and follow by lowercase letters.
* Use snake_case convention (i.e., separate words by underscores, look like a snake). Ex: gross_profit instead of grossProfit or GrossProfit.
* Should be meaningful and easy to remember. Ex: interest_rate instead of r or ir.
* Should have a reasonable length. Ex: sales_apr instead of sales_data_for_april
* Avoid names of popular functions and modules. Ex: avoid print, math, or collections.

* Control flow refers to control structures such as if-statements and for-loops that control the order in which code is executed.

#### `if` statements

In Python, an if statement is used to execute a block of code only if a specified condition is true. The condition is a boolean expression that is evaluated to either True or False. If the condition is True, the code block inside the if statement is executed. If the condition is False, the code block is skipped and the program continues with the next statement after the if block.

````python
if 5 + 5 == 10:
    print("This is True")
````

````python
x = 5
if x == 1:
    print(f"{x} = 1")
elif x == 2:
    print(f"{x} = 2")
elif x == 3:
    print(f"{x} = 3")
else:
    print(f"{x} > 3")
````

#### `for` loop

A for loop in Python is used to iterate over a sequence of elements, such as a list, tuple, or string. The sequence is the collection of elements over which we want to iterate, and element is a temporary variable that takes on the value of each element in the sequence, one at a time. The code block inside the loop is executed once for each element in the sequence, with the element variable taking on a new value for each iteration.

````python
for flower in ['tulip', 'rose', 'cucumber']:
    print(f"Is a {flower} a flower?")
````

````python
for i in range(5):
    print(i)
````

#### `while` loop

A while loop in Python repeatedly executes a block of code as long as the specified condition is true. The condition is a boolean expression that is evaluated at the beginning of each iteration. If the condition is true, the code block is executed, and the loop continues. If the condition is false, the loop is exited and the program continues with the next statement after the loop.

````python
# Print the first 5 positive integers using a while loop
i = 1  # Initialize the loop variable
while i <= 5:  # Set the loop condition
    print(i)  # Print the current value of i
    i += 1  # Increment the loop variable
````

````python
# Breaking a loop (see: https://docs.python.org/3/tutorial/controlflow.html)
i = 0
while True:  # This runs for a very long time ...
    print(i)
    i += 1
    if i >= 3:  # ... unless you break it.
        break
````

Arguments in a function can be optional, which requires setting a default value (can be `None`). Consider, for example, a function that calculates a future value considering different [compounding periods](https://en.wikipedia.org/wiki/Compound_interest) $n$.

$\text{FV} = \text{PV} \times (1+\frac{r}{n})^{nt}$

Compared to calculate_future_value() above, the function below has:
* an optional 'compounding' argument
* type hints
* a docstring

````python
def calculate_future_value(present_value:float, interest_rate:float, maturity:float, compounding:int=1) -> float:
    """Function to calculate a future value.

    Parameters
    ----------
    present_value : float
        DESCRIPTION.
    interest_rate : float
        DESCRIPTION.
    maturity : float
        DESCRIPTION.
    compounding : int, optional
        DESCRIPTION. The default is 1.

    Returns
    -------
    float
        DESCRIPTION.


    """
    return present_value * (1 + interest_rate / compounding) ** (maturity * compounding)

print(f"With annual compounding:       EUR {calculate_future_value(1000, 0.1, 1):,.2f}")
print(f"With semi-annual compounding:  EUR {calculate_future_value(1000, 0.1, 1, 2):,.2f}")
print(f"With daily compounding:        EUR {calculate_future_value(1000, 0.1, 1, 365):,.2f}")
````

## NumPy

NumPy is a well-known library utilized for storing numerical arrays and carrying out computations on them. This not only allows for more concise code but also enhances its efficiency as most NumPy functions are designed in C for faster processing.

See:
* https://numpy.org/doc/1.24/

Arrays are the central data structure in NumPy, and they are used to store homogeneous data. A NumPy array is similar to a list in Python, but it is more efficient in handling large amounts of data. In addition, NumPy arrays are highly optimized for numerical computations and support a wide range of mathematical operations.



To use NumPy in Python, you need to import the NumPy library using the import statement. Here is an example of importing NumPy and creating a simple array:

````python
import numpy as np

# create a one-dimensional array
arr = np.array([1, 2, 3, 4, 5])
print(arr)
````

NumPy provides several functions to create arrays of different shapes and sizes. For example, you can create a two-dimensional array using the `np.array()` function with a list of lists:

````python
# create a two-dimensional array
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2d)
````

You can also create arrays of zeros or ones using the `np.zeros()` and `np.ones()` functions, respectively:

````python
# create an array of zeros
zeros_arr = np.zeros((2, 3))
print(zeros_arr)

# create an array of ones
ones_arr = np.ones((2, 3))
print(ones_arr)
````

NumPy provides many built-in functions for mathematical operations, such as `np.sum()`, `np.mean()`, `np.std()`, and `np.dot()`. These functions can be used to perform mathematical operations on arrays:

````python
# sum of all elements in an array
arr_sum = np.sum(arr)
print(arr_sum)

# mean of all elements in an array
arr_mean = np.mean(arr)
print(arr_mean)

# standard deviation of all elements in an array
arr_std = np.std(arr)
print(arr_std)

# dot product of two arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
arr_dot = np.dot(arr1, arr2)
print(arr_dot)
````

````python
# generate some sample data
x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# create the figure and subplots
fig, ax1 = plt.subplots()

# create a second y-axis
ax2 = ax1.twinx()

# plot the data on the first y-axis
ax1.plot(x, y1, color='blue')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# plot the data on the second y-axis
ax2.plot(x, y2, color='red')
ax2.set_ylabel('cos(x)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# set the title of the plot
plt.title('Two y-axes example')

# Prevent the automatic display (I only need this for my slides)
plt.close(fig)
````

# Agentic coding

## Introduction

Agentic coding means using large language models (LLMs) and coding assistants not just for autocomplete, but as tools that can **reason, plan, write, debug, and revise code**. In practice, this means that you describe a task in natural language and the agent helps you turn it into working code, documentation, tests, or explanations.

For this course, agentic coding is useful because it can speed up routine tasks such as:
* explaining unfamiliar code,
* writing boilerplate,
* debugging errors,
* improving readability,
* and generating first drafts of analyses.

That said, an agent is **not a substitute for understanding**. It can produce convincing but wrong answers, misuse libraries, overlook assumptions, or generate code that runs without actually solving the problem correctly. Hence, in this class, we treat coding agents as **assistants, not authorities**.

Our goal is therefore simple: use agents to become **faster and more productive**, while still checking the logic, the data, and the output like a financial economist should.

## Which coding agents exist?

There are many coding agents, but for this course we focus on three practical options:

* **GitHub Copilot**: tightly integrated into GitHub and common editors such as Visual Studio Code. It is a convenient choice for code completion, chat, and coding support inside your development workflow.
* **Codex**: OpenAI's coding agent, designed to help write code, understand codebases, edit files, and assist with software development tasks.
* **Claude**: Anthropic's coding assistant, which can work across files, read codebases, run commands, and support more agentic development workflows.

A useful practical note for students: **verified students on GitHub Education can get GitHub Copilot Student at no cost**.

## How to use coding agents well

Coding agents are most useful when you give them **clear guardrails** and force them to work in **small, verifiable steps**.

A good workflow looks like this:

* **Set guardrails**: clearly state the goal, the available data, the relevant libraries, and any constraints. For example: keep it simple, use pandas, do not change unrelated code, and explain assumptions.
* **Ask for a plan first**: before generating code, the agent should outline the steps it wants to take. This makes it easier to catch bad ideas early.
* **Work incrementally**: do not ask for a huge solution in one shot. Solve one sub-problem at a time.
* **Verify each change**: after every step, check whether the code runs and whether the output is actually correct.
* **Keep control**: do not blindly accept generated code. You remain responsible for the logic, the interpretation, and the final result.

In short: **guardrails, plan, incremental changes, verification**.

### AGENTS.md

When working with a coding agent, it is often useful to define persistent instructions in an `AGENTS.md` file. Think of this file as a **rulebook for the agent**.

It can specify, for example:

* the coding style to follow,
* which libraries should be preferred,
* which patterns should be avoided,
* how much explanation is desired,
* and what the agent must never change without permission.

This is useful because it creates **consistent guardrails** across tasks. Instead of repeating the same instructions again and again, you write them down once and let the agent follow them throughout the project.

Example `AGENTS.md`:

```text
# AGENTS.md

## Scope
This project is for teaching financial data analytics in Python.

## Environment
- Run Python code in the `edu` environment.
- Manage the `edu` environment with `conda` and [environment.yml](/Users/woebbeking/Dropbox/Academia/Lectures/MLU_FinancialDataAnalytics/FinancialDataAnalytics/environment.yml).
- If environment-specific commands are needed, prefer clear and standard commands.

## Coding rules
- Write Python code according to PEP 8.
- Keep solutions simple and readable.
- Prefer pandas, numpy, and matplotlib.
- Avoid unnecessary dependencies.
- Do not rewrite unrelated code.
- Do not add complex abstractions unless clearly needed.
- When printing numeric values inside strings, use explicit formatting such as `:,.4f` where appropriate.

## Workflow rules
- For all non-trivial tasks, read and maintain `PLANS.md` as a living execution plan.
- Propose a short plan before making larger changes.
- Work in small, incremental, verifiable steps.
- Explain important assumptions briefly.
- Show complete code blocks that can be copied and run.
- Update the README when setup steps, dependencies, usage, or workflow change.

## Safety checks
- Do not claim code works unless it is plausible and internally consistent.
- Flag uncertainties clearly.
- Prefer transparent and boring solutions over clever but fragile ones.
```

### PLANS.md

For larger tasks, coding agents should first write an **execution plan** in `PLANS.md` rather than jumping straight into implementation.

Unlike `AGENTS.md`, which contains persistent project rules, `PLANS.md` is a **living document**. It should be updated as progress is made, as new findings emerge, and as design decisions are taken.

A good plan should at least maintain:
* **Progress**: granular steps with current status,
* **Surprises & Discoveries**: unexpected findings, bugs, or insights,
* **Decision Log**: important decisions and their rationale,
* **Outcomes & Retrospective**: what was achieved, what remains, and lessons learned.

Study carefully: [Codex exec plans](https://developers.openai.com/cookbook/articles/codex_exec_plans)


Example `PLANS.md`:

```text
# Build a stock-return script for class use

This ExecPlan is a living document. Keep `Progress`, `Surprises & Discoveries`, `Decision Log`, and `Outcomes & Retrospective` up to date as work proceeds.

## Purpose / Big Picture
Create a simple Python script for class that loads stock price data, computes daily returns, and plots cumulative performance. The result should run in the `edu` environment, stay easy to explain, and fit the teaching style of this course.

## Progress
- [x] (2026-04-07 10:00Z) Defined the scope: script, not notebook.
- [x] (2026-04-07 10:10Z) Chose pandas and matplotlib.
- [ ] Implement data loading.
- [ ] Implement return calculation.
- [ ] Implement cumulative return plot.
- [ ] Update README with setup and usage instructions.

## Surprises & Discoveries
- Some stock series may have missing observations.
- Adjusted prices are more appropriate than raw closing prices for return calculations.
- Ticker symbols and download coverage may differ across data sources.

## Decision Log
- Decision: Build a script instead of a notebook.
  Rationale: Easier to rerun and reuse in class.
- Decision: Use matplotlib instead of an interactive plotting library.
  Rationale: Simpler, more stable, and sufficient for teaching.
- Decision: Update the README together with the code.
  Rationale: Setup and usage should stay documented.

## Outcomes & Retrospective
- Not started yet.

## Working rules
- Update this file after meaningful progress.
- Revise the plan if the implementation approach changes.
- Record unexpected findings in `Surprises & Discoveries`.
- Record important choices in `Decision Log`.
- Keep `Progress` granular and current.
```

## Common mistakes in agentic coding

Coding agents can make you much faster, but they do not remove your responsibility.

Use them to:
* write boilerplate code,
* explain unfamiliar syntax,
* suggest revisions,
* debug routine problems,
* and speed up documentation.

Do **not** trust them blindly:
* they can invent functions, files, or APIs,
* they can misunderstand the task,
* they can produce code that runs but is still wrong,
* and they often miss edge cases.

The rule for this course is simple:

> **Use agents for speed, not for trust.**

You are still responsible for:
* understanding the code,
* verifying the output,
* checking assumptions,
* and making sure the result actually answers the question.

> When submitting homework or cases, show that you have **verified the agent's work**. In particular, I want to see **incremental progress in Git** with **meaningful commit messages**. I want to see the path toward the solution, not just 1,000 lines of agent-generated code dumped into the repository at once. Invest time in the plan and execution. **Never blindly commit anything!**
